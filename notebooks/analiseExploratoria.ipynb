```python
import pandas as pd
```

```python

df = pd.read_csv('../data/vendas_grande.csv')
```

```python
# limpeza de dados
df['data'] = pd.to_datetime(df['data'])
df.fillna(0, inplace=True)
```

```python
# Análise Estatística
df['receita_total'] = df['valor'] * df['quantidade']
```

```python
# Analise por grupos
valorEstado = df.groupby('estado')['valor'].sum()
vendedorEstado = df.groupby('estado')['vendedor'].count()
```

```python
# Visualização
vendedorEstado.plot(kind='bar')
```

```python
valorEstado.plot(kind='pie')
```
